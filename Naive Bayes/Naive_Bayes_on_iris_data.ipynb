# Import Libaraires

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
```

Loading IRIS data

```python
import pandas as pd

# Load the Iris dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Display the first 5 entries
print(df.head())
```

Split into Train and Test

```python
X, y = data.data, data.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

Initialize, Fit and Predict

```python
# Initialize the Naive Bayes classifier
model = GaussianNB()

# Fit the model to the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

```

Calclualte Accuracy

```python
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

## Visualization

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming y_test and y_pred are your data
test_values = y_test
predicted_values = y_pred

# Generating data points for the x-axis
data_points = np.arange(len(test_values))

# Plotting the data
plt.plot(data_points, test_values, label='Test Values', marker='o')
plt.plot(data_points, predicted_values, label='Predicted Values', marker='x')

# Adding title and labels
plt.title('Test vs Predicted Values for Iris Dataset')
plt.xlabel('Data Points')
plt.ylabel('Values')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ... (Code for Naive Bayes classifier)

# Data for the plot
labels = ['Naive Bayes']
accuracy_values = [accuracy]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

# Create the plot
fig, ax = plt.subplots()
rects1 = ax.bar(x, accuracy_values, width, label='Accuracy')

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy')
ax.set_title('Naive Bayes Classifier Accuracy')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Show the plot
plt.show()
```
